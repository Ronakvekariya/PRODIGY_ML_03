```python
import pandas as pd
import numpy as np
import os
from tensorflow.keras.preprocessing import image
```

```python
def read_all_images(directory , images):
    if os.path.exists(directory):
        for filename in os.listdir(directory):
            if filename.lower().endswith(('png', 'jpg', 'jpeg', 'tiff', 'bmp', 'gif')):
                try:
                    image_path = os.path.join(directory, filename)
                    img = image.load_img(image_path , target_size=(128, 128, 1), color_mode='grayscale')
                    img = image.img_to_array(img)
                    img = img / 255
                    images.append(img)
                except (Exception, OSError) as e:
                    print("Error: Unable to process image file.")
                    continue
                    
    else:
        print(f"The directory {directory} does not exist.")
    return images
```

```python
dogs_image = []
cats_image = []

directory = './Dog/' 
read_all_images(directory , dogs_image)

directory = './Cat/' 
read_all_images(directory , cats_image)
```

```python
value = 'dog'
dogs_label = [value]*len(dogs_image)
value = 'cat'
cats_label = [value]*len(cats_image)
```

```python
image_list = [dogs_image + cats_image ]
image_array = np.array(image_list)
label_list = [dogs_label +  cats_label]
label_array = np.array(label_list)

```

```python
len(label_array[0])
```

```python
from sklearn.model_selection import train_test_split

# Assuming joint_array and label_array are your features and labels
x_train , x_val , train_out , val_out = train_test_split(image_array[0] , label_array[0] , test_size = 0.2 , train_size = 0.8 , random_state = 42)
```

```python
n_samples = x_train.shape[0]
X_reshaped = x_train.reshape(n_samples, -1)
```

```python
from sklearn import svm

clf = svm.SVC(kernel='rbf', C=1.0, random_state=42)
```

```python
clf.fit(X_reshaped,train_out)
print('The Model is trained well with the given images')
# model.best_params_ contains the best parameters obtained from GridSearchCV
```

```python
predict = clf.predict(x_val)
```
